## 3. Basic Math
Python follows the PEMDAS order.

```python
# Store the total number of apples as an integer
apples = 10
# Store the number of friends sharing the apples as an integer
friends = 4
# Divide apples by friends — Python returns a decimal (float) when the result isn't a whole number
print('Each friend gets:', apples / friends)
```

## 4. Python Lists
Lists store multiple items in square brackets `[]`. Remember: Python starts counting at **0**!

```python
# Create a list with three fruit names — lists are ordered collections of items
fruits = ['Apple', 'Banana', 'Cherry']
# Access the first item using index 0 — Python starts counting positions from 0
print('First fruit:', fruits[0])

# Add a new item to the end of the list using .append()
fruits.append('Date')
# Print the updated list to confirm 'Date' was added
print('Updated list:', fruits)
```
